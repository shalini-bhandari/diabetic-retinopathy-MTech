```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# --- 1. Configuration ---
# Define the path to your CSV and the directory to save results
CSV_PATH = 'C:/Users/I768800/Documents/MTech_Project/diabetic-retinopathy-MTech/data/full_dataset/train.csv'
RESULTS_DIR = 'results/'
os.makedirs(RESULTS_DIR, exist_ok=True)


# --- 2. Load and Prepare Data ---
# Load the main dataframe
df = pd.read_csv(CSV_PATH)

# Get the count of each class
class_counts = df['diagnosis'].value_counts().sort_index()

# Define human-readable labels for the classes
class_labels = {
    0: 'No DR',
    1: 'Mild',
    2: 'Moderate',
    3: 'Severe',
    4: 'Proliferative DR'
}
class_names = [class_labels[i] for i in class_counts.index]


# --- 3. Generate the Plot ---
# Set the style for a professional look
sns.set(style="whitegrid", font_scale=1.1)

# Create the figure
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x=class_names, y=class_counts.values, palette="viridis")

# Add annotations (the exact count on top of each bar)
for index, value in enumerate(class_counts.values):
    plt.text(index, value + 20, str(value), ha='center', va='bottom', fontsize=12)

# Add titles and labels
plt.title('Class Distribution of the APTOS 2019 Dataset', fontsize=16, weight='bold')
plt.xlabel('Diabetic Retinopathy Severity Level', fontsize=14)
plt.ylabel('Number of Images', fontsize=14)
plt.xticks(rotation=15) # Slightly rotate x-axis labels for better readability

# Ensure everything fits and save the figure
plt.tight_layout()
plt.savefig(os.path.join(RESULTS_DIR, 'class_distribution.png'), dpi=300)

# Display the plot
plt.show()
```
